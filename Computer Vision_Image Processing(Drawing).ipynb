```python
import cv2
```

```python
# use OpenCV to load the image
eddycute = cv2.imread(r"C:\Users\user\Pictures\DCIM\Facebook\FB_IMG_1638875955059.jpg")
eddycute
```

```python
type(eddycute)
```

```python
# Dimensionality
eddycute.shape
```

```python
# Plot the image - OPENCV uses colour channel BGR(Blue, Green, Red), instead of RGB
plt.imshow(eddycute)
```

```python
# Convert Image from BGR to RGB
eddiecute = cv2.cvtColor(eddycute, cv2.COLOR_BGR2RGB)
plt.imshow(eddiecute)
```

```python
# Resizing the image - (Width, Height)
plt.imshow(cv2.resize(eddiecute, (320,480)))
```

```python
# Assign resized image to a new variable
img = cv2.resize(eddiecute, (320,480))

# Convert image from RGB to BGR
img1 = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

# Export the image
cv2.imwrite(r"C:\Users\user\Pictures\eddiecute_prac.jpg", img1)
```

```python
# Resize through %
width = 0.5
height = 0.5

eddiecute_r = cv2.resize(eddiecute, (0,0), eddiecute, width, height)
plt.imshow(eddiecute_r)
```

```python
# Flipping the image (0 - Flips horizontal axis)
eddiecute_f = cv2.flip(eddiecute, 0)
plt.imshow(eddiecute_f)
```

### Draw on Image

```python
# Create a blank Canva
blank = np.zeros(shape=(600,600,3), dtype=np.int16)
plt.imshow(blank)
```

```python
# Draw a rectangle - pt1(Top Left Corner), pt2(Bottom Left Corner)
cv2.rectangle(blank, pt1=(200,200), pt2=(400,400), color=(0,255,0), thickness=3)
plt.imshow(blank)
```

```python
# Draw a rectangle - pt1(Top Left Corner), pt2(Bottom Left Corner)
cv2.rectangle(blank, pt1=(250,250), pt2=(350,350), color=(0,0,255), thickness=3)
plt.imshow(blank)
```

```python
# Drawing a circle
cv2.circle(blank, center=(300,300), radius = 30, color=(255,0,0), thickness=-1)
plt.imshow(blank)
```

```python
# Drawing a circle
cv2.circle(blank, center=(500,100), radius = 50, color=(0,255,0), thickness=-1)
plt.imshow(blank)
```

```python
# Drawing a circle
cv2.circle(blank, center=(300,300), radius = 200, color=(0,255,0), thickness=8)
plt.imshow(blank)
```
